```python
import pandas as pd

import numpy as np
```

```python
raw_df = pd.read_csv("~/Desktop/thesis/experiment/data/Adams_experiment_cleaned_filtered.csv")
n = len(raw_df["subject"].unique())
months = raw_df.groupby("word").agg({"s2_value": "mean"}).reset_index()
```

```python
# Output: last, subject.id, value, order

df = pd.DataFrame({"subject.id": range(n)})
df = df.merge(months, how="cross").groupby("subject.id").sample(frac=1).reset_index(drop=True)
df["order"] = df.groupby("subject.id").cumcount()+1
df = df.rename(columns={"s2_value": "value"})
df.head(16)
```

```python
def generate_data(stop_proba_func):
    """
    Options
    stop_proba_func: Takes in df of subject's 12 months and must return 12 probabilities
    """

    def random_stopped(df):
        stop_proba = stop_proba_func(df)
        last_proba = stop_proba[1:] + [1]
        df["last"] = [True if x == 1 else False for x in np.random.binomial([1]*12, last_proba)]
        for i, x in enumerate(df["last"]):
            if x:
                df = df.iloc[:i+1]
                break
        return df

    return df.groupby("subject.id").apply(random_stopped).reset_index(drop=True)
```
